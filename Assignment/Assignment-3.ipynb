```python
import seaborn as sns
iris = sns.load_dataset('iris')
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue ="species",
              height = 6).map(plt.scatter,
                              'sepal_length', 
                              'petal_length').add_legend()
 
```

```python
sns.boxplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,size=5)
```

```python
sns.relplot(data=iris,x='sepal_length',y='sepal_width',hue='petal_length')
```

```python
iris_feat = iris.iloc[:,:-1]
iris_species = iris.iloc[:,-1]
```

```python
columns = iris_feat.columns
x_data = range(0, iris.shape[0])
fig, ax = plt.subplots()
for column in columns:
    ax.plot(x_data, iris[column], label=column)
ax.set_title('Iris Dataset')
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.hist(iris_feat['sepal_length'])
ax.set_title('sepal_length')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

```python
df = iris.groupby('species')['sepal_length'].sum().to_frame().reset_index()
plt.bar(df['species'],df['sepal_length'],color = ['cornflowerblue','lightseagreen','steelblue']) 
plt.title('Bar Chart')
plt.xlabel('Species')
plt.ylabel('sepal_length') 
plt.show()
```
